```python
# Markov Decision Process (MDP representation)
states=[0,1,2]
actions=[0,1]

P={
    0:{0:[(1,1.0)], 1:[(2,1.0)]},
    1:{0:[(0,1.0)], 1:[(2,1.0)]},
    2:{0:[(2,1.0)], 1:[(2,1.0)]}
}

R={
    0: {0:5,1:10},
    1:{0:-1, 1:2},
    2:{0:0,1:0}
}

print("---------MDP Structure-------")
print("States",states)
print("Actions",actions)
print("Transitions",P)
print("Rewards",R)
```

```python
# Policy Evaluation (Iterative)
def policy_evaluation(policy,P,R,gamma=.9, theta=1e-5):
    V=np.zeroes(len(policy))
    while True:
        delta=0
        for s in range(len(policy)):
            v=V[s]
            a=policy[s]
            V[s]=R[s][a]+gamma*sum(prob*V[next_state] for next_state, prob in P[s][a])
            
            delta=max(delta, abs(v-V[s]))
        if (delta< theta):
            break
    return V

policy=[0,1,0]
V=policy_evaluation(policy,P,R)

print("\n-------Policy Evaluation--------")
for i,v in enumerate(V):
    print(f"State {i}: {v:.4f}")
```
