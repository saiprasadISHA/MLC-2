```python
import numpy as np;
#imputs
x=np.array([.5,.2,.1])

# wights
w=np.array([.4,-.7,.2])

y_sum=np.dot(x,w)
output=y_sum

print("Weighted Sum:",y_sum)
print("Output" ,output)
```

```python
def biological_neuron(signals, threshold=1.0):
    soma=sum(signals)
    if soma>=threshold:
        return 1    #neuron fires
    return 0

signals=[.3,.4,.5]
print("Neuron Output ",biological_neuron(signals))
```

```python
# 3) Artificial neuron wih weights+ activation

import numpy as np

def artificial_neuron(inputs, weights, bias=0):
    y_sum=np.dot(inputs,weights)+bias
    return y_sum

x=np.array([1,0,1])
w=np.array([.5,-.6,.2])
print("Summation Output ", artificial_neuron(x,w,bias=.1))
```

```python
import math

# Identity function
def identity(x):
    return x;

# Step/ Threshold Function
def step(x):
    return 1 if x>0 else 0

# ReLU Function
def threshold(x,theta=.5):
    return 1 if x>=theta else 0

# ReLU Function
def relu(x):
    return max(0,x)

# Sigmoid Function
def sigmoid(x, k=1):
    return 1/(1+math.exp(-k*x))

# Bipolar function
def bipolar_sigmoid(x):
    return (2/(1+math.exp(-x))-1)

# Hyperbolic Tangent
def tanh_n(x):
    return math.tanh(x)

# ---------Test Values------------
values=[-2,-1,0,.5,1,2]
print("Imput values: ",values)
print('\n--- Identity Function----')
for v in values:
    print(v, '->',step(v))

print('\n---Threshold Functon (theta=0.5)-----')
for v in values:
    print(v,'->', threshold(v))

print('\n----ReLU Function----')
for v in values:
    print(v,'->', relu(v))

print('\n-------Sigmoid Functon------ ')
for v in values:
    print(v,'->', round(sigmoid(v),4))

print('\n-------Bipolar Sigmoid Functon------ ')
for v in values:
    print(v,'->', round(sigmoid(v),4))

print('\n-------Theta Functon------ ')
for v in values:
    print(v,'->', round(tanh_n(v),4))
```
